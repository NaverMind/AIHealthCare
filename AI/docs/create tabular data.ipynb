### # 정답 label 붙이기

```python
label = [x for x in df_train.columns if "pts" not in x]
label
```

```python
label_3d = ["active", "type", "exercise", "value0", "value1", "value2", "value3", "value4", "condition0", "condition1", "condition2", "condition3", "condition4"]
label_2d = ["view1.active", "type", "type_info.exercise",  "value0", "value1", "value2", "value3", "value4", "condition0", "condition1", "condition2", "condition3", "condition4"]
```

```python
for i in range(len(label_2d)):
  df_train_3d[label_3d[i]] = df_train[label_2d[i]]
```

```python
for i in range(len(label_2d)):
  df_valid_3d[label_3d[i]] = df_valid[label_2d[i]]
```

```python
df_train_3d.to_csv("../gdrive/MyDrive/EDA/dataset/body_3d_add.csv")
df_valid_3d.to_csv("../gdrive/MyDrive/EDA/dataset/valid_body_3d_add.csv")
```
